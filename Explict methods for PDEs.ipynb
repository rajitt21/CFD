```python
# Runga Kutta Method or RK2
import numpy as np
phi = np.zeros(100)
phi[0] = 1
dt=0.1
for i in range(1,100):
  k1 = phi[i-1]*dt*(-1)
  k2 = phi[i-1]*(-dt + dt*dt)
  phi[i] = phi[i-1] + (k1+k2)/2

x = np.arange(0,10,0.1)
print(np.shape(phi))
print(np.shape(x))
import matplotlib.pyplot as plt
plt.plot(x,phi)
```

```python
# Runga Kutta Method or RK4
import numpy as np
phi = np.zeros(100)
phi[0] = 1
dt=0.1
for i in range(1,100):
  k1 = phi[i-1]*dt*(-1)
  k2 = phi[i-1]*(-dt + dt*dt/2)
  k3 = phi[i-1]*(-dt + dt*dt/2 - dt*dt*dt/4)
  k4 = phi[i-1]*(-dt + dt*dt - dt*dt*dt/2 + dt*dt*dt*dt/4)
  phi[i] = phi[i-1] + (k1+2*k2+2*k3+k4)/6

x = np.arange(0,10,0.1)
print(np.shape(phi))
print(np.shape(x))
import matplotlib.pyplot as plt
plt.plot(x,phi)
```

```python
# RKW3 Method
import numpy as np
phi = np.zeros(100)
phi[0] = 1
dt = 0.1
for i in range(1,100):
  phi_str = phi[i-1]
  k1 = -phi_str
  phi_str = phi_str + k1*dt/3
  k1 = k1*5/9 - phi_str
  phi_str = phi_str + k1*dt*15/16
  k1 = -k1*153/128 - phi_str
  phi[i] = phi_str + k1*dt*8/15

x = np.arange(0,10,0.1)
print(np.shape(phi))
print(np.shape(x))
import matplotlib.pyplot as plt
plt.plot(x,phi)
```
